```python
# function use to impute_nan values with mean/median/mode
def impute_nan(DataFrame, ColumnName, ImputeValue):
    DataFrame[ColumnName] =   DataFrame[ColumnName].fillna(ImputeValue)
```

```python
# Call function to impute median value
median = Data.Age.median()
impute_nan(Data, 'Age', median)
```

```python
# Handle categorical missing data
```

```python
#1. Function to replace NAN values with mode value
def impute_nan_most_frequent_category(DataFrame,ColName):
    # .mode()[0] - gives first category name
     most_frequent_category=DataFrame[ColName].mode()[0]
    
    # replace nan values with most occured category
     DataFrame[ColName] = DataFrame[ColName]
     DataFrame[ColName].fillna(most_frequent_category,inplace=True)
```

```python
#2. Call function to impute most occured category
for Columns in ['Embarked','Cabin_Serial','Cabin']:
    impute_nan_most_frequent_category(Data,Columns)
```

# 3. Count / Frequency Encoding

```python
Data = Dataset.copy()
```

```python
# 1. Create list of all categorical columns
CatCols = Data.select_dtypes(include=['object', 'category']).columns.tolist()
```

```python
CatCols
```

```python
# 1. Create Dictionary with key as category name and value with count of number of categories i.e frequency of that category in each categorical column

Pclass_Dict = Data['Pclass'].value_counts()
Salutation_Dict = Data['Salutation'].value_counts()
Sex_Dict = Data['Sex'].value_counts()
Embarked_Dict = Data['Embarked'].value_counts()
Cabin_Serial_Dict = Data['Cabin_Serial'].value_counts()
Cabin_Dict = Data['Cabin'].value_counts()


```

```python
Salutation_Dict
```

```python
# 2. Create new column which act as weight for that category and map with its respective dictionary.

Data['Encoded_Pclass'] = Data['Pclass'].map(Pclass_Dict)
Data['Salutation_Dict'] = Data['Salutation'].map(Salutation_Dict)
Data['Sex_Dict'] = Data['Sex'].map(Sex_Dict)
Data['Embarked_Dict'] = Data['Embarked'].map(Embarked_Dict)
Data['Cabin_Serial_Dict'] = Data['Cabin_Serial'].map(Cabin_Serial_Dict)
Data['Cabin_Dict'] = Data['Cabin'].map(Cabin_Dict)

```

```python
# Display Result
Data[['Pclass','Encoded_Pclass','Salutation','Salutation_Dict','Sex','Sex_Dict','Embarked','Embarked_Dict','Cabin_Serial','Cabin_Serial_Dict','Cabin','Cabin_Dict']].head(10)


```

```python
# 3. Drop Orginal Columns
Data = Data.drop(['Pclass','Salutation','Sex','Embarked','Cabin_Serial','Cabin'], axis = 1)
```

# Target/Guided Encoding

```python
Data = Dataset.copy()
```

```python
# 1. Take first character of cabin and store into same column.
Data['Cabin'] = Data['Cabin'].astype(str).str[0]
```

```python
Data['Cabin']
```

```python
#2. Calculate probability of each category based on target value
Data.groupby(['Cabin'])['Survived'].mean()
```

```python
#3. Create a list with index sorted in ascending order of probabilites
Encoded_Lables = Data.groupby(['Cabin'])['Survived'].mean().sort_values().index
```

```python
Encoded_Lables
```

```python
#4. Create dictionary with key as categories in cabin and values as probaibility ranking

Encoded_Lables_Ranks = { k:i for i, k in enumerate(Encoded_Lables, 0) }
```

```python
Encoded_Lables_Ranks
```

```python
#5. Create new column with map cabin values with dictionary probability ranking
Data['Cabin_Encoded'] = Data['Cabin'].map(Encoded_Lables_Ranks)
```

```python
Data[['Cabin','Cabin_Encoded']].head(20)
```

```python
# 6. Delete original cabin column.

Data = Data.drop('Cabin', axis = 1)
```

# Probability Ratio Encoding

```python
Data = Dataset.copy()
```

```python
# 1. Replace original cabin value with the first character of the cabin name.
Data['Cabin']  = Data['Cabin'].astype(str).str[0]
```

```python
# 2. Find percentage (%) of people survived in particular cabin and convert it into dataframe.

Probability_Survived = Data.groupby(['Cabin'])['Survived'].mean()
Probability_Survived = pd.DataFrame(Probability_Survived)
```

```python
Probability_Survived
```

```python
# 3. Create new column into probability survived dataframe with probability of people getting dead in particular cabin.
Probability_Survived['Died'] = 1 - Probability_Survived['Survived']
```

```python
Probability_Survived
```

```python
# 4. Create one more new column into probability survived dataframe i.e ratio of survived and died probablity

Probability_Survived['Prob_Ratio'] = Probability_Survived['Survived'] / Probability_Survived['Died']
```

```python
# 5. Create dictionary with probability ratio column

Encode_Prob_Ratio = Probability_Survived['Prob_Ratio'].to_dict()
```

```python
# 6. Create new column in Data and replace by maping cabin column categories with its encoded probablity ratio dictionary

Data['Encode_Prob_Ratio'] = Data['Cabin'].map(Encode_Prob_Ratio)
```

```python
Data[['Cabin','Encode_Prob_Ratio']].head(10)
```

```python
# 7. Drop original cabin column
Data = Data.drop('Cabin', axis = 1)
```
